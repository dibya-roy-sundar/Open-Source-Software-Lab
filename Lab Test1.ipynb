```python
import numpy as np
df=np.round(np.random.random((1440,1))*100)
print(df)
```

```python
def calculate_avg():
    a=df.reshape(24,60)
    d=np.average(a,axis=1).reshape(1,24)
    return d
    


avg=calculate_avg()
print(avg[0])
```

```python
import matplotlib.pyplot as plt
from scipy import signal

x=np.linspace(0,1440,1440);

plt.plot(x,df)

plt.title("aqi data v/s minutes")
plt.xlabel("minutes")
plt.ylabel("aqi data")
plt.show()
```

```python
z=np.linspace(0,23,24)
y=avg[0]
plt.scatter(z,y,s=2.5,color='r')
plt.title("avg aqi vs hours")
plt.xlabel("hour")
plt.ylabel("avg aqi")
plt.show()
```
